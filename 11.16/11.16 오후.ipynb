```python
import pandas as pd
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
treemodel=DecisionTreeClassifier()
```

```python
treemodel.get_params()
```

```python
param_tree={'criterion':['gini','entropy'],'max_depth':range(1,11)}
```

```python
ss=ShuffleSplit(test_size=0.25,n_splits=10)
```

```python
grid_model=GridSearchCV(treemodel,return_train_score=True, param_grid=param_tree, cv=ss, verbose=1,scoring='f1_macro')
```

```python
grid_model.fit(X,Y)
```

```python
grid_model.best_score_
```

```python
grid_model.best_estimator_.get_params()
```

```python
cross_val_score(grid_model,X,Y,cv=3)
```

```python
preferred_params=grid_model.best_estimator_.get_params()
```

```python
final_tree=DecisionTreeClassifier(**preferred_params)
```

```python
final_tree.fit(X,Y)
```

```python
final_tree.score(X,Y)
```
