```python
p = []

# 1. 영화 페이지 link를 드라이버가 겟해서 하나씩 창을 열게 만들기
for li in atag_link[:2]:
    for j in li:
        driver.get((j))
        time.sleep(2)
    # 2.영화 예매 페이지 안에서 리뷰 긁어오기(리뷰자, 리뷰내용, 리뷰날짜)
        pages = driver.find_elements(By.CSS_SELECTOR, '#paging_point > li > a') # 페이지 번호 가져오기
       # 리스트로 저장된 element속성의 pages들을 하나씩 빼서 페이지 번호를 텍스트로 가져와 페이지 번호 리스트(p)에 넣기
        for page in pages:
            
            p.append(page.text)
     # 3. 리뷰 페이지 넘겨가며 읽어오기
            for i in p:    # 페이지 번호 리스트(p)에서 하나씩 뽑아와서 리뷰 페이지 넘어가게 만들기
                pc = driver.find_elements(By.CSS_SELECTOR, '#paging_point > li:nth-child(' + i + ') > a')
                for k in pc: # 뽑아온 리뷰 페이지 번호의 경로가 저장된 pc에서 하나씩 경로를 뽑아서
                    k.click() # 하나씩 뽑아온 경로 k를 click해서 이동하게 만들기
                    time.sleep(0.5) # 데이터를 뽑아올 시간을 주기 위해 딜레이 시간 주기

                        # 4. 영화 예매 페이지 안 리뷰 페이지에서 리뷰 긁어오기
                    
```
